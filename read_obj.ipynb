```

import subprocess
subprocess.run(['pip', 'install', 'openpyxl', 'xlrd', '-q'], capture_output=True)

import pandas as pd
import os

paths = [
    '/workspace/Obiettivi_Segreti_Linea_Rossa.xlsx',
    '/workspace/uploaded_files/Obiettivi_Segreti_Linea_Rossa.xlsx',
]

for p in paths:
    if os.path.exists(p):
        print(f"\n=== {p} ===")
        xl = pd.ExcelFile(p)
        print(f"Fogli: {xl.sheet_names}")
        for sheet in xl.sheet_names:
            df = pd.read_excel(p, sheet_name=sheet)
            print(f"\n--- Foglio: {sheet} --- ({len(df)} righe, {len(df.columns)} colonne)")
            print(f"Colonne: {list(df.columns)}")
            print(df.to_string())
```
